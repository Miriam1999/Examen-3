## Romeo y Julieta (Modelo 1)

### Problema 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import matplotlib as mpl
%matplotlib inline
def armonico(y,B):
    a=1
    b=1
    y1=y[0]
    y2=y[1]
    return [a*y2,-b*y1]
y0=[5,5]                                 
tt=np.linspace(0,50,500)
yy=odeint(armonico,y0,tt)
yy
```

### Problema 2

```python
y1=yy[:,0]
y2=yy[:,1]
label_size=14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size
plt.figure(figsize=(8,8))
plt.plot(tt,y1,"b",label="Amor-odio de Romeo a Julieta")
plt.plot(tt,y2,"r",label="Amor-odio de Julieta a Romeo")
plt.legend(loc="best")
plt.xlabel("Tiempo")
plt.grid()
plt.show()
```

La **gráfica del amor/odio de Romeo hacia Julieta** parece ser una función seno y la **gráfica de amor/odio de Julieta hacia Romeo** parece ser una función coseno. Ambas funciones trigonométricas tienen la misma amplitud e inician en la misma ordenada al origen. Cuando Romeo es indiferente, Julieta lo odia o lo ama intercaladamente; y también funciona igual al revés.

### Problema 3

```python
w0=np.sqrt(1/1)
plt.figure(figsize=(8,8))
plt.scatter(y1,y2/w0,c=y2,cmap="viridis")
plt.xlabel("Amor-odio de Romeo a Julieta")
plt.ylabel("Amor-odio de Julieta a Romeo")
plt.grid()
plt.show()
```

**El amor entre Romeo y Julieta no es sano** porque como se puede apreciar en la gráfica de fase, mientras uno ama al otro, los sentimientos del otro no son recíprocos en la mayoría de los casos. Es por eso que la figura que se forma al graficar los sentimientos de ambos es un círculo. Cuando Romeo está en una etapa de máximo amor (o odio) hacia Julieta, ella es indiferente porque en ambos casos, su eje se encuentra en 0. De igual manera, cuando los sentimientos de Julieta hacia Romeo son extremos (lo ama mucho o lo odia mucho), el eje de Romeo se encuentra en 0. **Los únicos momentos en donde sus sentimientos son iguales son en las coordenadas de fase (-5,-5) y (5,5)**.

## Romeo y Julieta (Modelo 2)

### Problema 1

```python
def armonico(y,B):
    a=2
    b=1
    y1=y[0]
    y2=y[1]
    return [-a*y1+b*y2,b*y1-a*y2]
y0=[-2,1]                                 
tt=np.linspace(0,50,500)
yy=odeint(armonico,y0,tt)
yy
```

### Problema 2

```python
y1=yy[:,0]
y2=yy[:,1]
label_size=14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size
plt.figure(figsize=(8,8))
plt.plot(tt,y1,"b",label="Amor-odio de Romeo a Julieta")
plt.plot(tt,y2,"r",label="Amor-odio de Julieta a Romeo")
plt.legend(loc="best")
plt.xlabel("Tiempo")
plt.grid()
plt.show()
```

El amor-odio de ambos parecen tener una forma de función exponencial porque hay una asíntota muy evidente en y=0. Esta asíntota aparece en todos los casos de condiciones iniciales propuestos, que son (2,1), (-2,1), (-2,-1) y (2,-1). No importa cuales sean sus sentimientos hacia el otro al principio, en cierto punto (aproximadamente cuando t=5), **las emociones de ambos se estabilizan en dicha asíntota cuando el tiempo tiende a infinito**, lo que quieren decir que terminan siendo indiferentes.

### Problema 3

```python
w0=np.sqrt(2/1)
plt.figure(figsize=(8,8))
plt.scatter(y1,y2/w0,c=y2,cmap="viridis")
plt.xlabel("Amor-odio de Romeo a Julieta")
plt.ylabel("Amor-odio de Julieta a Romeo")
plt.grid()
plt.show()
```

Tampoco considero sano este tipo de modelo (aunque cabe decir que es mejor que el anterior) porque aunque sus sentimientos hacia el otro se terminan estabilizando, esa estabilidad aparece porque ya hubo mucha intensidad entre ellos y terminan siendo indiferentes. En la gráfica de fase se puede ver que al principio **ella lo ama y él la odia**, pero en lo que **Romeo está en el proceso de dejar de odiar a Julieta y empezar a amarla, el amor de ella va disminuyendo** (incluso por un momento empieza a odiarlo también), por lo que Romeo nunca llega a amarla (ya que los sentimientos de ambos se quedan estancados en la indiferencia).

### Problema 4

 - [5 puntos] Cambiando los parámetros $a=2,b=1$ y partiendo de alguna condicion inicial, proponga un modelo de amor que considere sano y explique cómo devería ser este amor
 - [5 puntos] Graficar el retrato de fase, es decir, $x_2$ vs. $x_1$.

```python
def armonico(y,B):
    a=0.1
    b=0.1
    y1=y[0]
    y2=y[1]
    return [-a*y1+b*y2,b*y1-a*y2]
y0=[1,9]                                 
tt=np.linspace(0,50,500)
yy=odeint(armonico,y0,tt)
yy
```

```python
y1=yy[:,0]
y2=yy[:,1]
label_size=14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size
plt.figure(figsize=(8,8))
plt.plot(tt,y1,"b",label="Amor-odio de Romeo a Julieta")
plt.plot(tt,y2,"r",label="Amor-odio de Julieta a Romeo")
plt.legend(loc="best")
plt.xlabel("Tiempo")
plt.grid()
plt.show()
```

```python
w0=np.sqrt(0.1/0.1)
plt.figure(figsize=(8,8))
plt.scatter(y1,y2/w0,c=y2,cmap="viridis")
plt.xlabel("Amor-odio de Romeo a Julieta")
plt.ylabel("Amor-odio de Julieta a Romeo")
plt.grid()
plt.show()
```

Considero que este modelo es el más óptimo porque como en el modelo 2, **es bueno que haya cierta estabilidad** pero no que ésta se encuentre en 0. **La estabilidad debe encontrarse en un valor positivo que no sea 0 para que ambos sientan amor uno por el otro**. La condición inicial que puse es de 1,9 porque el amor de Romeo es de nivel 1, es decir, Julieta es algo así como solamente su crush. Sin embargo el nivel de amor de Julieta al principio es de 9, uno muy elevado. Con el paso del tiempo el amor de Julieta pasa de ser una obsesión enfermiza a un amor sano, llegando al nivel 5. Esto pasa mientras a Romeo le empieza a gustar cada vez más Julieta y también llega al nivel 5. Así, el amor de ambos se estabiliza en 5, convirtiéndose en una relación sana (considerando que ese es el mejor nivel de amor para una pareja sana). La gráfica de fases muestra cómo los puntos se concentran en la coordenada (5,5).
