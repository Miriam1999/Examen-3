### Problema 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
def armonico(y,B):
    a=1
    b=1
    y1=y[0]
    y2=y[1]
    return [a*y2,-b*y1]
y0=[5,5]                                 
tt=np.linspace(0,50,500)
yy=odeint(armonico,y0,tt)
yy
```

```python
y1=yy[:,0]
y2=yy[:,1]
import matplotlib as mpl
label_size=14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size
plt.figure(figsize=(8,6))
plt.plot(tt,y1,"b",label="Amor-odio de Romeo a Julieta")
plt.plot(tt,y2,"r",label="Amor-odio de Julieta a Romeo")
plt.legend(loc="best")
plt.xlabel("Tiempo")
plt.grid()
plt.show()
```

La **gráfica del amor/odio de Romeo hacia Julieta** parece ser una función seno y la **gráfica de amor/odio de Julieta hacia Romeo** parece ser una función coseno. Ambas funciones trigonométricas tienen la misma amplitud e inician en la misma ordenada al origen. Cuando Romeo es indiferente, Julieta lo odia o lo ama intercaladamente; y también funciona igual al revés.

# Nota: 

Se pueden considerar romances más generales con el siguiente modelo
\begin{align}
\frac{dx_1}{dt}&=a\;x_1+b\;x_2\\
\frac{dx_2}{dt}&=c\;x_1+d\;x_2,
\end{align}
donde $a$, $b$, $c$ y $d$ constantes que pueden ser tanto positivas como negativas.

Por ejemplo, si $a>0$ y $b>0$, podríamos llamar a Romeo un "amante apasionado". O, si $a<0$ y $b>0$, podríamos llamar a Romeo un "amante cauteloso".
